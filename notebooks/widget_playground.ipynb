# Widget Playground

This notebook demonstrates a few ipywidgets building blocks you can extend for image-to-image translation experiments.

```python
import ipywidgets as widgets
from IPython.display import display
```

## Interactive slider
Move the slider to see live updates. Replace the body of `preview` with image processing logic later.

```python
def preview(mix: float):
    if mix < 0.33:
        status = 'line art focus'
    elif mix < 0.66:
        status = 'balanced blend'
    else:
        status = 'color focus'
    return f"Mix: {mix:.2f} → {status}"

widgets.interact(preview, mix=widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.05, description='Blend'))
```

## Widget layout
Combine widgets to build simple control panels.

```python
style_toggle = widgets.ToggleButtons(
    options=['Sketch', 'Flat Color', 'Shaded'],
    description='Mode:',
    button_style='',
)
noise_slider = widgets.FloatSlider(value=0.1, min=0.0, max=0.5, step=0.05, description='Noise')
options = widgets.Checkbox(value=True, description='Preserve text bubbles')

control_panel = widgets.VBox([style_toggle, noise_slider, options])
display(control_panel)
```
